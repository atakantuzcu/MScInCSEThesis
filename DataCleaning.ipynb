```python
#Loading the necessary libraries.
import pandas as pd
```

```python
#Reading dataset to pandas.DataFrame
df = pd.read_csv("rawMatchData.csv")
df.shape
```

```python
#Checking the dataset for the Null values
df.isnull().sum()
```

```python
#Dropping the missing values from the dataframe.
df = df.dropna()
```

```python
#Verifying if there are any more missing data.
df.isnull().sum().sum()
```

```python
#Concatanate the platformId column with gameId as Id column
df["id"]=df["platformId"]+df["gameId"].astype(str)
df = df.drop(df.columns[[0, 1]], axis=1)
df.insert(0, 'id', df.pop('id'))
```

```python
#This data was 1 player match so we dropped it
id_value_counts = pd.DataFrame(df["id"].value_counts())
id_value_counts[id_value_counts["count"]==1]
```

```python
index = df[df["id"]=="KR6109288002"].index[0]
df = df.drop([index])
```

```python
df
```

```python
#Observing the win-lose distribution of the dataset
df.win=[ 1 if i==True else 0 for i in df.win] #assigning 1 to victory, 0 to defeat.
df["win"].value_counts()
```

```python
#Filtering the data and groupby it to get the team based data.
df_teambased = df.groupby(["id","win"]).sum().reset_index()
df_teambased
```

```python
df_playerbased = df
```

```python
#To output the clean data as TeamBased and PlayerBased version.
df_teambased.to_csv("cleanDataTeamBased.csv",index=False)
df_playerbased.to_csv("cleanDataPlayerBased.csv",index=False)
```
